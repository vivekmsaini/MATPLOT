pie plot

```python
population = [700,560,450,510,300]
country = ['india','china','japan','bangladesh','pakistan']

plt.pie(population,labels=country)
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['india','china','japan','bangladesh','pakistan']

plt.pie(population,labels=country,autopct='%1.2f%%')    ### with the help of autopct function we consider the data with the percentage 
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['india','china','japan','bangladesh','pakistan']

plt.pie(population,labels=country,autopct='%1.2f%%',wedgeprops={'ec':'k'})  
plt.title('country population distribution in oie ploty')    ## add a title with the help of title function
plt.show()
```

```python
country = ['india','china','japan','bangladesh','pakistan']
exp = [0,0.3,0,0,0.4]
## explode function is distribute the graph data 
plt.pie(population,labels=country,autopct='%1.2f%%',wedgeprops={'ec':'k'},explode=exp)  
plt.title('country population distribution in oie ploty')   
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['India','china','japan','bangladesh','nepal']
exp = [0,0.3,0,0,0.2] 
plt.pie(population,labels = country, autopct = '%1.2f%%',wedgeprops={'ec':'k','ls':'dashed'},explode=exp,shadow = True)
plt.title('Country population distribution in pie plot')
plt.show()
```
